```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Store filepath in a variable
file_one = "census_data_by_zip.csv"
file_two = "large_brewery_file.csv"
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
file_one_df = pd.read_csv(file_one, encoding="ISO-8859-1")
file_one_df["zip_code"]=file_one_df["zip_code"].astype(str)
# Show just the header
file_one_df.head()
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
file_two_df = pd.read_csv(file_two, encoding="ISO-8859-1")
file_two_df.dropna(subset=["lat","long"])
# Show just the header
file_two_df.head()
#print(file_two_df.columns)
```

```python
# Merge two dataframes
merge_table = pd.merge(file_one_df, file_two_df, on="zip_code")
merge_table.to_csv("Joined_File_Zip_Code.csv", index=False, header=True)
merge_table
```
