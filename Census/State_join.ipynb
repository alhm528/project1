```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Store filepath in a variable
file_one = "census_data.csv"
file_two = "breweries.csv"
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
file_one_df = pd.read_csv(file_one, encoding="ISO-8859-1")
# Show just the header
file_one_df.head()
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
file_two_df = pd.read_csv(file_two, encoding="ISO-8859-1")
# Show just the header
file_two_df.head()
```

```python
# Merge two dataframes using an inner join
merge_table = pd.merge(file_one_df, file_two_df, on="Name")
merge_table.to_csv("Joined_File.csv", index=False, header=True)
merge_table
```

```python
states = merge_table["Name"]
number_breweries = merge_table["State"].value_counts()

# Configure plot and ticks
plt.figure(figsize=(20,3))
number_breweries.plot(kind="bar", facecolor="seagreen")

# Set textual properties
plt.title("Number of Breweries by State")
plt.ylabel("Number of Breweries")
plt.xlabel("State")

# Show plot
plt.show()
plt.tight_layout()
```

```python
num_brew = number_breweries
grouped = merge_table.groupby("Name")
median_age = grouped["Median Age"].mean()

plt.scatter(number_breweries, median_age, marker="o", facecolors="violet", edgecolors="black")

plt.ylim(34, 44)

# Set the upper and lower limits of our x axis
plt.xlim(0, 50)

# Create a title, x label, and y label for our chart
plt.title("Number of Breweries vs. Median Age")
plt.xlabel("Number of Breweries")
plt.ylabel("Median Age (in years)")

plt.show
```

```python
grouped = merge_table.groupby("Name")
population = grouped["Population"].mean()

plt.scatter(number_breweries, population, marker="o", facecolors="orange", edgecolors="black")

plt.ylim(50000, 40000000)

# Set the upper and lower limits of our x axis
plt.xlim(-5, 50)

# Create a title, x label, and y label for our chart
plt.title("Number of Breweries vs. Population")
plt.xlabel("Number of Breweries")
plt.ylabel("Population")

plt.show
```

```python
grouped = merge_table.groupby("State")
median_age = grouped["Median Age"].mean()

plt.scatter(median_age, number_breweries, marker="*", facecolors="violet", edgecolors="black")

plt.ylim(-5, 50)

# Set the upper and lower limits of our x axis
plt.xlim(34, 44)

# Create a tit;e, x label, and y label for our chart
plt.title("Age vs. Number of Breweries")
plt.xlabel("Median Age")
plt.ylabel("Breweries in State")

plt.show
```

```python
grouped = merge_table.groupby("State")
income = grouped["Household Income"].mean()

plt.scatter(number_breweries, income, marker="x", facecolors="deeppink", edgecolors="black")

plt.ylim(35000, 80000)

# Set the upper and lower limits of our x axis
plt.xlim(0, 50)

# Create a title, x label, and y label for our chart
plt.title("Number of Breweries vs. Income")
plt.xlabel("Number of Breweries")
plt.ylabel("Income")

plt.show
```
