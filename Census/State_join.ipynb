```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Store filepath in a variable
file_one = "OUtput/census_data.csv"
file_two = "large_brewery_file.csv"
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
file_one_df = pd.read_csv(file_one, encoding="ISO-8859-1")
# Show just the header
file_one_df.head()
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
file_two_df = pd.read_csv(file_two, encoding="ISO-8859-1")
# Show just the header
file_two_df.head()
```

```python
print(file_two_df.columns)
```

```python
# Merge two dataframes using an inner join
merge_table = pd.merge(file_one_df, file_two_df, on="Name")
merge_table.to_csv("Output/Joined_File.csv", index=False, header=True)
merge_table
```

```python
states = merge_table["Name"]
number_breweries = merge_table["Name"].value_counts()

# Configure plot and ticks
plt.figure(figsize=(20,3))
number_breweries.plot(kind="bar", facecolor="seagreen")

# Set textual properties
plt.title("Number of Breweries by State")
plt.ylabel("Number of Breweries")
plt.xlabel("State")

# Show plot

plt.savefig("Output/breweries_by_state.png")

plt.show()
plt.tight_layout()
```

```python
number_breweries = merge_table["Name"].value_counts()
grouped = merge_table.groupby("Name")
median_age = grouped["Median Age"].mean()

median_age = pd.DataFrame({"Number of Breweries": number_breweries,
                       "Median Age": median_age
                      })

median_age.to_csv("Output/median_age.csv", index=False, header=True)

median_age
```

```python
number_breweries = merge_table["Name"].value_counts()
grouped = merge_table.groupby("Name")
population = grouped["Population"].mean()

population = pd.DataFrame({"Number of Breweries": number_breweries,
                       "Population": population
                      })

population.to_csv("Output/population.csv", index=False, header=True)

population
```

```python
number_breweries = merge_table["Name"].value_counts()
grouped = merge_table.groupby("Name")
income = grouped["Household Income"].mean()

income = pd.DataFrame({"Number of Breweries": number_breweries,
                       "Household Income": income
                      })

income.to_csv("Output/income.csv", index=False, header=True)

income
```
